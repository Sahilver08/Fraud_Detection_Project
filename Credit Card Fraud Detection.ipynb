Train-Test Split

```python
# Features and target
X = df.drop('Class', axis=1)
y = df['Class']

# Split the data (stratify to maintain class balance)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
# Create SMOTE object
smote = SMOTE(random_state=42)

# Fit and resample training data only
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Check class balance
print("Before SMOTE:", y_train.value_counts())
print("After SMOTE:", y_train_resampled.value_counts())
```

```python
# Initialize XGBoost model
model = XGBClassifier(
    n_estimators=100,
    max_depth=5,
    learning_rate=0.1,
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)

# Train the model on the balanced training data
model.fit(X_train_resampled, y_train_resampled)
print("Model trained successfully.")
```

```python
# Predict on test set
y_pred = model.predict(X_test)

# Evaluation
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# ROC AUC Score
auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
print("ROC AUC Score:", auc)
```

```python
# Get prediction probabilities for fraud (class 1)
y_proba = model.predict_proba(X_test)[:, 1]

# Compute false positive rate and true positive rate
fpr, tpr, thresholds = roc_curve(y_test, y_proba)

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label="XGBoost (AUC = {:.4f})".format(roc_auc_score(y_test, y_proba)), color='blue')
plt.plot([0, 1], [0, 1], linestyle="--", color='gray')  # Diagonal line
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.grid(True)
plt.show()
```
